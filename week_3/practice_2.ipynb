```python
from tkinter import *
from tkinter import messagebox

def button_click():
    msgbox.showinfo("Info", "Welcome to COS102 GUI App" );

    result = msgbox.askyesno("Confirmation", "Do you want to continue");

root = Tk();
root.title("Home page");
root.geometry("300x100");

label = Label(root, text="Hello Friend \n");
label.pack();

button = Button(root, text="Click Me!!!!", command=button_click);
button.pack();

button.config(fg="pink", bg="white");

root.mainloop();
```

```python
from tkinter import *
from tkinter import messagebox
#from PIL import Image, ImageTk

def welcome_message(username):
    window = Toplevel(root);
    window.title("Admin Box");
    window.geometry("500x200");
    
    label_1 = Label(window, text=f"Welcome {username}\n");
    label_1.pack();

    label_2 = Label(window, text=f"This is Python GUI with Tkinter");
    label_2.pack();

#root.mainloop();

def submit():
    username = username_entry.get();
    password = password_entry.get();

    if username == "Mary" and password == "cos102":
        welcome_message(username);
    else:
        messagebox.showerror("Login", "Invalid username or password");

root = Tk();
root.title("Login Form");
root.geometry("500x200");

username_label = Label(root, text="Username");
username_label.pack();
username_entry = Entry(root);
username_entry.pack();

password_label = Label(root, text="Password");
password_label.pack();
password_entry = Entry(root);
password_entry.pack();

submit_button = Button(root, text="submit", command=submit);
submit_button.pack();

root.mainloop();
```
