```python
def printme(str):
    print(str);
    return

printme("I'm first call to user defined function!");
printme("I'm second call to the same function!");
```

```python
#pass by reference
def changeme(mylist):
    mylist.append([1,2,3,4]);
    print("Values inside the function: ", mylist);
    return

mylist = [10,20,30];
changeme(mylist);
print("Values outside the function: ", mylist);
```

```python
def changeme(mylist):
    # the argument passed will be overwritten locally
    mylist = [1,2,3,4];
    print("Values inside the function:", mylist);
    return

mylist = [10,20,30];
changeme(mylist);
print("Values outside the function: ", mylist)
```

```python
#required args
def printme(str):
    print(str);
    return

# this will not work, the required arg was not passed
printme();

# this is a keyword argument
# the value is declared directly inside the func call
printme(str = "My string");
```

```python
def printinfo(name, age):
    print("Name ", name);
    print("Age ", age);
    return

printinfo(age= 50, name= "milkie");
```

```python
#default arguments
def printinfo(name, age = 35):
    print("Name ", name);
    print("Age", age);
    return

printinfo(age=50, name= "milkie");

#the age of milkie will default to 35 if no arg is passed
printinfo(name = "milkie");
```

```python
# variable arguments
def printinfo(arg1, *vartuple):
    print("Output is: ");
    print(arg1);

    for var in vartuple:
        print(var);
        return

printinfo(10);
printinfo(70, 60, 50); 
#notice how 50 didn't print? remember 70 is now arg1, position 1!
#that forloop is gonna start from idx[1] not [0]! 
printinfo(70,60);
```

```python
#total is a global var
total = 50;

def sum(arg1, arg2):
    total = arg1 + arg2; #this var is only local!
    print("Inside the function, total: ", total);
    return total

sum(10,20);
print("Outside the function (globally), total: ", total);
```
