```python
import csv
from tkinter import *
from tkinter import messagebox, scrolledtext

root = Tk();
root.title("Employee login");
root.geometry("400x300");

def show_department_members(department):
    results_box = scrolledtext.ScrolledText(root, width=45, height=10, state=DISABLED)
    results_box.pack(pady=10)

    results_box.config(state=NORMAL)
    results_box.delete("1.0", END)

    #FIND ONLY MEMBERS IN THE USER'S DEPARTMENT
    members = [row[1] for row in employees if row[2].lower() == department.lower()]
    results_box.insert(END, f"Members in {department}:\n\n")

    # loop through and write members to box
    for member in members:
        results_box.insert(END, f"  • {member}\n")
    results_box.config(state=DISABLED)

# Validation function
def validate_entry():
    username = name_entry.get().strip()
    department = department_entry.get().strip()
    
    # Check if any row matches both username and department
    found = False
    for row in employees:
        if row[1].lower() == username.lower() and row[2].lower() == department.lower():
            found = True
            break
    
    if found:
        messagebox.showinfo("Welcome!", f"Welcome back, {username}")
        show_department_members(department);
    else:
        messagebox.showerror("Error", f"{username} does not exist in employee records")


employees = [];
with open("generated_names.csv", newline='') as employee_records:
    reader = csv.reader(employee_records);
    for row in reader:
        employees.append(row);  # each row is a list of values

    Label(root, text="Welcome user!").pack();

    Label(root, text="Enter your employee username:").pack();

    name_entry = Entry(root);
    name_entry.pack();

    Label(root, text="Enter your Department:").pack();

    department_entry = Entry(root);
    department_entry.pack();

    submit = Button(root, text="Submit", command=validate_entry)
    submit.pack();

root.mainloop();

def collect_input():
    name = input("Please enter your employee username!").lower().strip();
    department = input("Now, enter your department name").lower().strip();
    validate_entry();



# Button
submit = Button(root, text="Submit", command=validate_entry)
submit.pack();
#just ignore the error this line is gonna give you

root.mainloop();

```
