```python
import tkinter as tk
from tkinter import messagebox
from tkinter import scrolledtext
```

```python
departments = [
    'Retail Banking',
    'Global Banking',
    'Commercial Banking'
]
```

```python
class Zenith():
    
    def unique_services(self):
        return []
    def mutual_services(self):
        return [
            'Lines of credit',
            'Investment management and accounts',
            'Insurance'
        ]
```

```python
class retail_banking(Zenith):
    def unique_services(self) -> list:
        return [
            'Retirement and education accounts',
            'Loans and mortages',
            'Checking and saving'
        ]
    
class global_banking(Zenith):
    def unique_services(self):
        return [
            'Multi-currency management',
            'Foreign currency accounts',
            'Foreign currency credit cards',
            'Transborder advisory services',
            'Liquidity management'
        ]

class commercial_banking(Zenith):
    def unique_services(self):
        return [
            'Advisory services'
        ]
```

```python
class GUI:

    def __init__(self, root):
        self.root = root
        self.root.title = ("Zenith Banking Services")
        self.root.geometry("400x400")
        self.root.configure(bg="pink")

        tk.Label(
            root,
            text = "Zenith Bank Services",
            bg = "pink"
        ).pack()

        # Collect Employee name
        tk.Label(root, text="Enter your full name:", bg="pink").pack()
        self.name_entry = tk.Entry(root)
        self.name_entry.pack()

        # Collect Employee dept
        tk.Label(root, text="Enter your department name:", bg="pink").pack()
        self.department_entry = tk.Entry(root)
        self.department_entry.pack()

        # Button
        tk.Button(
            root, 
            text="View Available Services", 
            command=self.check_employee,
            bg = "#fc4e68"
            ).pack(pady=15)
        
        self.output = scrolledtext.ScrolledText(self.root, width=40, height=10)
        self.output.pack(pady=10)
        
    def check_employee(self):
        name = self.name_entry.get().strip()
        department = self.department_entry.get().strip().lower()

        if department == 'retail banking':
            employee = retail_banking()
        elif department == 'global banking':
            employee = global_banking()
        elif department == 'commercial banking':
            employee = commercial_banking()
        else:
            messagebox.showerror('Error', 'The specified department does not exist')
            return
        
        services = employee.mutual_services() + employee.unique_services()

        self.output.delete("1.0", tk.END)

        for service in services:
            self.output.insert(tk.END, f"- {service}\n")


        
```

```python
if __name__ == '__main__':
    root = tk.Tk()
    GUI(root)
    root.mainloop()
```
