```python
import csv
import tkinter as tk
from tkinter import messagebox
```

```python
class DeliveryCalculator:
    def __init__(self, root):
        self.root = root
        self.root.title("Delivery Fee Calculator")
        self.root.geometry("350x250")

        # Title
        tk.Label(root, text="Delivery Service", font=("Arial", 14, "bold")).pack(pady=10)

        # Location input
        tk.Label(root, text="Enter Location (PAU / Epe):").pack()
        self.location_entry = tk.Entry(root)
        self.location_entry.pack()

        # Weight input
        tk.Label(root, text="Enter Package Weight (kg):").pack()
        self.weight_entry = tk.Entry(root)
        self.weight_entry.pack()

        # Button
        tk.Button(root, text="Calculate Fee", command=self.calculate_fee).pack(pady=15)

        # Result label
        self.result_label = tk.Label(root, text="", font=("Arial", 12))
        self.result_label.pack()

    def calculate_fee(self):
        location = self.location_entry.get().strip().lower()

        try:
            weight = float(self.weight_entry.get())
        except ValueError:
            messagebox.showerror("Error", "Please enter a valid number for weight")
            return

        cost = 0

        if location == "pau":
            if weight >= 10:
                cost = 2000
            else:
                cost = 1500

        elif location == "epe":
            if weight >= 10:
                cost = 5000
            else:
                cost = 4000
        else:
            messagebox.showerror("Error", "Location must be PAU or Epe")
            return

        self.result_label.config(text=f"Delivery Fee: ₦{cost}")
```

```python
# Run app
if __name__ == "__main__":
    root = tk.Tk() #root is an object of the Tk class!
    #root2 = tk.Tk()
    app = DeliveryCalculator(root)
    #app2 = DeliveryCalculator(root2)

    print(type(app));
    root.mainloop()
```
