```python
import csv
import tkinter as tk
from tkinter import messagebox, scrolledtext
import random
from datetime import datetime
```

### Read each task from tasks.csv

```python
tasks = [];
with open("tasks.csv", newline='') as task_names:
    reader = csv.reader(task_names);
    for row in reader:
        tasks.append(row);  # each row is a list of values
print(tasks);
```

### Now read employee names from csv file

```python
employees = [];
with open("employees.csv", newline='', encoding='utf-8-sig') as employee_names:
    reader = csv.reader(employee_names)
    next(reader) #skip the header row
    for row in reader:
        employees.append([cell.strip() for cell in row])  # strip each cell
print(employees);
```

```python
class Employees:
    def __init__(self, root):
        self.root = root
        self.root.title("Employee Attendance")
        self.root.geometry("350x250")
        self.root.configure(bg="pink")

        # Title
        tk.Label(
                root, 
                text="Employee Attendance", 
                font=("Arial", 14, "bold"),
                bg= "pink"
                ).pack(pady=10)

        # Collect Employee name
        tk.Label(root, text="Enter your full name:", bg="pink").pack()
        self.name_entry = tk.Entry(root)
        self.name_entry.pack()

        # Button
        tk.Button(
            root, 
            text="Take Attendance", 
            command=self.check_employee,
            bg = "#fc4e68"
            ).pack(pady=15)

        # Result label
        self.result_label = tk.Label(root, text="", font=("Arial", 12), bg = "pink")
        self.result_label.pack()
    
    def check_employee(self):
        full_name = self.name_entry.get().strip().lower()
        #unformatted employee name:
        self.employee_name = self.name_entry.get().strip();
        print(full_name);

        found = False
        found = any(row[0].lower() == full_name for row in employees)

        if found:
            messagebox.showinfo("Welcome!", f"Welcome back, {self.employee_name}")
            self.take_attendance()
        else:
            self.refuse_access();
            print(f"Name used is {full_name}")
        
    def refuse_access(self):
        messagebox.showerror("Error", f"{self.employee_name} does not exist in records");
        return
    
    def take_attendance(self):
        current_time = datetime.now();
        self.formatted_time = current_time.strftime("%Y-%m-%d %H:%M:%S");
        print(self.formatted_time);
        messagebox.showinfo("Entry added successfully", 
                            f"Your attendance entry has been recorded at\n{self.formatted_time}");

        self.assign_task()
    
    def assign_task(self):
        #generate a random integer between 1 and 5
        #each int will represent an idx in the tasks list
        task_index = random.randint(1,len(tasks) -1);
        print(tasks[task_index]);
        self.random_task = tasks[task_index];
        messagebox.showinfo("Today's task", 
                            f"Your task for today is: {self.random_task}");

        with open("attendance.csv", "a", newline="") as file:
            writer = csv.writer(file)
            writer.writerow([
                            self.employee_name, 
                            self.formatted_time, 
                            self.random_task
                                            ])
```

```python
if __name__ == "__main__":
    root = tk.Tk();
    app = Employees(root)

    root.mainloop()
```
