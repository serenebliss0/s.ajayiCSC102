```python
print("Welcome to my Simple Interest calculator.");
print("Please input a value for your principal");

principal = float(input());

print("Now input the rate");

rate = float(input());

print("Now input the time (in years)");

time = float(input());

amount = principal * (1 + (rate/100) * time);

print("Your final amount is:", amount);
```
