```python
def get_float(msg):
    while True:
        try:
            return float(input(msg));
        except ValueError:
            print("Invalid value entered. Please try again.");
    
def simple_interest():
    while True:
        print("Welcome to my Simple Interest calculator.");

        principal = get_float("First enter your principal amount");

        rate = get_float("Now input the rate in percentage");

        time = get_float("Now input the time (in years)");

        amount = principal * (1 + (rate/100) * time);

        print("Your final amount is:", amount);

        user_opt = input("Do you want to run again? Type y or n");
        if user_opt != "y":
            break;

def compound_interest():
    
    while True:
        print("Compound Interest calculator!");

        principal = get_float("Please enter your principal amount");

        rate = get_float("Now enter your rate");
    
        n_years = get_float("Now enter the number of years");

        t_times = get_float("Finally enter how many times the interest is compounded per year (n)");

        rate = rate / 100;
        
        amount = principal * ((1 + (rate / t_times)) ** (t_times * n_years))
        print("The interest is ", amount);

        print("Would you like to run the program again?\nType y or n");
        close_option = input();

        if close_option != "y" :
            break;


def annuity():

    while True :
        print("Annuity plan calculator!");

        pmt = get_float("To get started please enter a PMT (Payments per period) value");

        rate = get_float("Now enter your rate");

        n_value = get_float("Now enter the number of times the interest is compounded");

        num_years = get_float("Lastly enter the number of years");

        rate = rate /100;
        amount = pmt * ((((1 + (rate/num_years))** (num_years*n_value)) - 1)/ (rate/num_years));
        print("The amount is ", amount);

        return_option = input("Do you want to rerun the program?\nType y or n");

        if return_option != "y":
           break;

print("Please choose an option!");
print("1. Simple Interest Calculator\n2. Compound Interest Calculator\n3. Annuity Calculator");

user_option = int(input());

match user_option:
    case 1:
        simple_interest();
    case 2:
        compound_interest();
    case 3:
        annuity();
```
