## Project 1C
### Algorithm to swap age of two names

```
REM "Program to swap the ages of two people"

PRINT "Enter the age of person 1"
INPUT age_1

PRINT "Now enter the age of person 2"
INPUT age_2

LET temp = age_1
LET age_1 = age_2
LET age_2 = temp

PRINT "The age of person one after swap is"
PRINT age_1

PRINT "The age of person two after swap is"
PRINT age_2
```
