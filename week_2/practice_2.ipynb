```python
city1 = input("Enter name of city 1");
city2 = input("Enter name of city 2");

temp = city1;
city1 = city2;
city2 = temp;

print(f"The name of City 1 after swap is {city1}");
print(f"The name of City 2 after swapping is {city2}");
```

```python
num = float(input("Enter a number"));

if num > 0:
    print("This is a positive number");
elif num == 0:
    print("This is zero");
else:
    print("This is a complex number");
```
