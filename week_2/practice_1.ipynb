```python
str = input("Enter your state of origin");

print("Your state of origin is: " , str);
print("The first character is: " , str[0]);
print("The characters starting from 3rd to 5th are: " , str[2:5]);
print("The string starting from the 3rd char is: ", str[2:]);
print("State of origin two times", str * 2);
```

```python
m = float(input("Enter mass in kilograms: "))

c = 299892458;

energy = m * c ** 2;

print(f"The energy equivalent to {m} kilograms is {energy} joules");
```

```python
list = ['Anaconda', 786, 2.23, 'Jupyter', 70.2 ];
shortlist = [321, 'Python'];

print(list);
print(list[0]);
print(list[1:3]);
print(list[2:]);
print(shortlist * 2);
print(list + shortlist);
```

```python
tuple = ("Ekiti", 750, 'Oshogbo', 250, "Akure", 500);
s_tuple = ("Abeokuta", 300, "Ogbomoso");

print(tuple);

print(tuple[-1]) # use -1 to print the last element

print(tuple[2:4]);

print(tuple[3:]);

print(s_tuple * 3);

print(tuple + s_tuple);
```

```python
game_1 = 2;
game_2 = 4;
print(bool(game_1 == game_2));

val = None; # return false as val is none
print(bool(val));

num = () # false as val in an empty sequence
print(bool(num));

age = True;
print(bool(age));
```

```python
grade = int(70);
gpa = int(4.9) #grade will truncate to 4 (like in rust))
cgpa = int("4");

print(f"Grade = {grade}");
print(f"GPA = {gpa}");
print(f"CGPA = {cgpa}");
```
