```python
def get_int(message:str):
    while True:
        try:
            return int(input(message));
        except ValueError:
            print("Input a valid number!");


def main():
    print("Welcome client!");
    staff_age = get_int("How old are you?\n");
    years_of_experience = get_int("How many years of experience do you have?\n");

    tax_revenue = 0;

    if years_of_experience > 25 and staff_age >= 55:
        tax_revenue = 5_600_000;
    elif years_of_experience > 20 and staff_age >= 45:
        tax_revenue = 4_480_000;
    elif years_of_experience > 10 and staff_age >=35:
        tax_revenue = 1_500_000;
    else:
        tax_revenue = 550_000;

    print(f"Your ATR with age {staff_age} and experience of {years_of_experience} years, your ATR is:\n{tax_revenue}");


if __name__ == "__main__":
    main();
```
