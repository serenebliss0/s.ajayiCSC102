```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.read_csv("titanic.csv");

#most values here are null so i dropped them
data.drop(['Cabin'], axis=1, inplace=True);

#i filled the empty age values with the median age instead
data['Age'] = data['Age'].fillna(data['Age'].median())
data.info()
```

# Survivors by Gender

Most survivors were women, while more than 50% of men did not survive.

```python
#create a column to count how many survived or didn't survive
data['Survived'] = data['Survived'].map({1: 'Survived', 0: 'Did Not Survive'});

sns.countplot(data, x=data['Sex'], hue="Survived", hue_order=['Did Not Survive', 'Survived']);
plt.title("Number of Survivors by Gender");
plt.show();
```

# Number of people that survived with siblings or spouses

People who had only one sibling/spouse had a higher chance of surviving

```python
#create a column to count how many survived or didn't survive
#data['Survived'] = data['Survived'].map({1: 'Survived', 0: 'Did Not Survive'});

sns.countplot(data, x=data['SibSp'], hue="Survived", hue_order=['Did Not Survive', 'Survived']);
plt.title("Number of Survivors with Siblings or Spouses");
plt.xlabel("Number of Siblings or Spouses");
plt.ylabel("Count")
plt.show();
```

# Survivors by Ticket Class

People in 1st Class survived 

```python

sns.countplot(data, x=data['Pclass'], hue="Survived", hue_order=['Did Not Survive', 'Survived']);
plt.title("Number of Survivors based on Ticket Class");
plt.xlabel("Ticket Class");
plt.ylabel("Count")
plt.show();
```
